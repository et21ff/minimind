# PyTorch 基础操作练习
本笔记本记录了常用的 Tensor 操作，包括条件选择、序列生成、线性代数和维度变换。

### 1. 条件选择：`torch.where`
`torch.where(condition, x, y)`：如果满足条件则取 x 的值，否则取 y 的值。

```python
import torch

x = torch.tensor([1, 2, 3, 4, 5])
y = torch.tensor([10, 20, 30, 40, 50])

result = torch.where(x > 3, x, y)
print(f"torch.where 结果: {result}") # 预期: [10, 20, 30, 4, 5]
```

### 2. 序列生成：`torch.arange`
生成等差数列。参数为 `(start, end, step)`，不包含 `end`。

```python
t1 = torch.arange(0, 10, 2)
print(f"正向序列: {t1}") # [0, 2, 4, 6, 8]

t2 = torch.arange(5, 0, -1)
print(f"反向序列: {t2}") # [5, 4, 3, 2, 1]
```

### 3. 线性代数：向量外积 `torch.outer`

```python
v1 = torch.tensor([1, 2])
v2 = torch.tensor([4, 5,6])
print(f"外积结果:\n{torch.outer(v1, v2)}")
```

### 4. 张量拼接：`torch.cat`
在现有维度上连接张量。

```python
t1 = torch.ones(2, 2, 3)
t2 = torch.ones(2, 2, 3)

print(f"原始形状: {t1.shape}")
print(f"Dim 0 拼接: {torch.cat((t1, t2), dim=0).shape}")
print(f"Dim 1 拼接: {torch.cat((t1, t2), dim=1).shape}")
print(f"最后维度拼接: {torch.cat((t1, t2), dim=-1).shape}")
```

### 5. 维度扩展：`unsqueeze`

```python
t1 = torch.tensor([[1, 2, 3],[1,2,3]])
print(f"原始形状: {t1.shape}")
print(f"Unsqueeze(0) 形状: {t1.unsqueeze(0).shape}")
```
